## Run a solver and plot data

The beauty is that we can simply switch the style to generate the right plot for the occasion!

```python
#| label: fig-plot
#| fig-cap: "Caption"
# import matplotlib
# styles can be found here https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
# print(matplotlib.style.available)
import matplotlib.pyplot as plt
import sample_code as code
# plt.style.use('seaborn-v0_8-talk')
from matplotlibstyle import *

fig, ax = plt.subplots(2, constrained_layout=True)
x, y = code.solver()
code.plot(ax[0], x, y, label='sin(x)')
code.plot(ax[0], x, 2*y, label='2*sin(x)')
code.plot(ax[1], x, 3*y, label='3*sin(x)')
ax[0].legend()
ax[1].legend()
plt.show()
```
